```python
import networkx as nx
nx.draw(nx.barbell_graph(5, 1))
```

```python
from networkx.algorithms.community import LFR_benchmark_graph
n = 250
tau1 = 3
tau2 = 1.5
mu = 0.1
G = LFR_benchmark_graph(n, tau1, tau2, mu, average_degree=5, min_community=20, seed=10)

from pylab import *
plt.figure(figsize=[20,20])
pos=nx.spring_layout(G)

subplot(2,2,1)
nx.draw_spectral(G, node_size = 8, with_labels = False)
subplot(2,2,2)
nx.draw(G, pos, node_size = 8, with_labels = False)
subplot(2,2,3)
nx.draw_spectral(G, node_size = 8, with_labels = True)
subplot(2,2,4)
nx.draw(G, pos, node_size = 8, with_labels = True)
```

```python
nx.get_node_attributes(G, all)
nx.get_edge_attributes(G, all)
```

```python
G1 = G
```

```python
import colorcet as cc
import matplotlib
plt.figure(figsize=[30,20])


subplot(2,3,1)
nx.draw(G1, pos, with_labels = True, node_color = cc.bjy[0:250],
        node_size = np.array(list(nx.betweenness_centrality(G1).values()))*1000)
subplot(2,3,2)
nx.draw(G1, pos, with_labels = False, node_color = cc.bjy[0:250],
        node_size = np.array(list(nx.betweenness_centrality(G1).values()))*1000)
subplot(2,3,3)
nx.draw(G1, pos, with_labels = True, node_color = cc.bgyw[0:250],
        node_size = np.array(list(nx.degree(G1))))
subplot(2,3,4)
nx.draw(G1, pos, with_labels = True, node_color = cc.isolum[0:250],
        node_size = np.array(list(nx.eigenvector_centrality(G1).values()))*1000)
subplot(2,3,5)
nx.draw(G1, pos, with_labels = False, node_color = cc.isolum[0:250],
        node_size = np.array(list(nx.eigenvector_centrality(G1).values()))*1000)

plt.savefig('Graph0.PDF', format = "PDF")
plt.show()
plt.close()
```
